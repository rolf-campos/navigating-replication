### Requirements
The code requires that the folder `data` is populated with estimation results. The code is othwerwise self-contained.


```python
import os
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## LP and VAR

```python
# Read results for the VAR and LP
df = pd.read_excel(os.path.join("data", "IRF_VAR_LP_paper.xlsx"), sheet_name="LP")
```

```python
# Figure 3

y = 100 * df['IRmean'].values
lb = 100 * df['IRinf90'].values
ub = 100 * df['IRsup90'].values
lb2 = 100 * df['IRinf68'].values
ub2 = 100 * df['IRsup68'].values
t = list(range(len(y)))
fig, ax = plt.subplots(figsize=(6, 4))

ax.plot(t, y, color='k', lw=1)
ax.fill_between(t, lb, ub, color='#fcdede', label="90% confidence interval")
ax.fill_between(t, lb2, ub2, color='#ffa8a8', label="68% confidence interval")

ax.axhline(y=0.0, color='k', linewidth=0.5, linestyle=':')


plt.xlabel('Quarter')
plt.ylabel('Basis points') 
ax.set_ylim(bottom=-8, top=15)
ax.legend(loc='lower left', frameon=False)
plt.tight_layout()

#plt.title(r"$r^\ast - \overline{r}$")
savename = "LP.pdf"
plt.savefig(os.path.join("results", savename), format="pdf")
plt.show()
```
